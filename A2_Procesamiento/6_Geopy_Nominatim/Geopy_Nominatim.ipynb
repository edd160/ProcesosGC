Dirección a coordenadas

```python
import pandas as pd
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent='cctmexico')
df = pd.DataFrame({'direcc':['2094 Valentine Avenue, Bronx, NY, 10457',
                             '1123 East Tremont Avenue, Bronx, NY, 10460',
                             '412 Macon Street Brooklyn, NY, 11233',
                             'Calle del Universo, 3, Valladolid',
                             '302 Juan de Montoro, Aguascalientes',
                             '419 Calle 117, Los Héroes, Mérida']})
```

```python
import time
start = time.time()

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

df['location'] = df['direcc'].apply(geocode)
df['coordenadas'] = df['location'].apply(lambda x:(x.latitude, x.longitude))

end = time.time()
elapsed = end-start
print(df)
print(str(elapsed)+"seconds")
```

Coordenadas a Dirección

```python
from geopy.extra.rate_limiter import RateLimiter

reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)
df['Time'] = df['coordenadas'].apply(reverse)
df['Dirección'] = df['Time'].apply(lambda x: (x.address))
df
```
