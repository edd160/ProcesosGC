B) Se obtiene la lista de enlaces donde se encuentra el contenido de interés

```python
url = 'https://meganews.mx/seccion/yucatan/'
medio = 'meganews'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

for link in soup.find_all('a'):
    enlace = link.get('href')
    if '#main'not in enlace and enlace != '/' and 'mailto' not in enlace and 'youtube' not in enlace:
        #print(enlace)
        new_row = pd.DataFrame({'enlace':enlace, 'medio':medio}, index=[0])
        df_noticias = pd.concat([new_row,df_noticias.loc[:]]).reset_index(drop=True)
```

```python
url = 'https://meganews.mx/seccion/yucatan/'
medio = 'meganews'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

for link in soup.find_all('a'):
    enlace = link.get('href')
    if '#main'not in enlace and enlace != '/' and 'mailto' not in enlace and 'youtube' not in enlace and 'https://meganews.mx/yucatan/' in enlace:
        #print(enlace)
        new_row = pd.DataFrame({'enlace':enlace, 'medio':medio}, index=[0])
        df_noticias = pd.concat([new_row,df_noticias.loc[:]]).reset_index(drop=True)
```

Ahora vamos a extraer todos los archivos de los enlaces de la muestra

```python
i=1
for index, row in df_muestra.iterrows():
    url = row['enlace']
    r = requests.get(url, allow_redirects=True)
    open('noticia'+str(i)+'.html', 'wb').write(r.content)
    i += 1
```
